The yields of apple trees in an orchard have been recorded in the file `yield.csv`. Each observation is recorded by weighing apples from trees (in pounds) and adding their weights. There are 5000 observations in total for this data. 

## Load, visualize and give general comments about the dataset

Use pandas for loading and inspecting the data.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Read the yield data as a dataframe
df = pd.read_csv('yield.csv')
df
```

```python
# Create a plot
plt.figure(figsize = (12,8))
ax = sns.distplot(df,
             hist = True, 
             hist_kws = {"linewidth": 2,
                        "edgecolor" :'red',
                        "alpha": 0.4, 
                        "color":  "w",
                        "label": "Histogram",
                        "density": 1    
                       },
             kde = True,
             kde_kws = {'linewidth': 3,
                        'color': "blue",
                        'alpha': 0.7,
                        'label':'Kernel Density Estimation Plot'
                       },
             fit = stats.norm,
             fit_kws = {'color' : 'green',
                        'label' : 'parametric fit',
                        'alpha': 0.7,
                        'linewidth':3,
                        'label': 'Normal PDF'}
                  )
      
ax.set(xlabel = 'Yeild per tree in pounds',
       ylabel='Proportion of trees',
       title= 'Yield Distribution'
      )
```

This data appears to be well approximated by a normal distribution. Mean and standard deviation are computed below. 

```python
print('Mean:', df['0'].mean())
print('Standard Deviation:', df['0'].std())
```

## Briefly explain what each value represents in this data set

The proportion of trees with yields in the given yield bin.

## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
mu = df['0'].mean()
sigma = df['0'].std()
print('99% of all trees have a yeild in the following interval.', f'[{round(mu-3*sigma,2)}, {round(mu+3*sigma,2)}]')
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
z = (35 - mu)/sigma
```

```python
# Interpret the result
print(f'A tree having a yeild of 35 pounds falls {round(abs(z),2)} standard deviations below the mean.')
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

# Interpret the z score
This tree has a fairly high yeild relitive to the average.

```python
# Calculate yield
print(f"This tree would yeild {round(sigma*1.85 + mu,2)} pounds of apples.")
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution here 
df['standardized'] = df['0'].apply(lambda x: (x-mu)/sigma)
```

```python
# Your observations
plt.figure(figsize = (12,8))
ax = sns.distplot(df.standardized,
             hist = True, 
             hist_kws = {"linewidth": 2,
                        "edgecolor" :'red',
                        "alpha": 0.4, 
                        "color":  "w",
                        "label": "Histogram",
                        "density": 1    
                       },
             kde = True,
             kde_kws = {'linewidth': 3,
                        'color': "blue",
                        'alpha': 0.7,
                        'label':'Kernel Density Estimation Plot'
                       },
             fit = stats.norm,
             fit_kws = {'color' : 'green',
                        'label' : 'parametric fit',
                        'alpha': 0.7,
                        'linewidth':3,
                        'label': 'Normal PDF'}
                  )
      
ax.set(xlabel = 'z-score of yeild per tree',
       ylabel='Proportion of trees',
       title= 'stadardized Yeild Distribution'
      )
```
